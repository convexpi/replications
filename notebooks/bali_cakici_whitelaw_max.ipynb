```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from replications import all_replications
r = all_replications()['bali_cakici_whitelaw_max']
rc = r.report_card()
print(rc['title'], '\u2014 verdict:', rc['verdict'])
print('IS Sharpe :', rc['in_sample_sharpe'])
print('OOS Sharpe:', rc['oos_sharpe'], ' (decay:', rc['sharpe_decay'], ')')
```

```python
s = r.build().dropna()
(1 + s).cumprod().plot(logy=True, figsize=(9,4), title=rc['title'] + ' (recomputed)')
plt.axvline(pd.Timestamp(f"{r.pub_year}-01-01"), color='crimson', ls='--', label='published')
plt.legend(); plt.tight_layout(); plt.show()
```
